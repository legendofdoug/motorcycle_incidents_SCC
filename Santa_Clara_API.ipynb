```python
MyAppToken = getpass.getpass('What if your App Token?')
```

```python

username = input ("What is your email?: ")
password = getpass.getpass ("What is your password")
```

```python
client = Socrata('data.sccgov.org',
                 MyAppToken,
                 username,
                 password)
```

```python
results = client.get("n9u6-aijz", limit=200000)
```

```python
results_df = pd.DataFrame.from_records(results)
```

```python
#Only Traffic parent incident
traffic_mask = results_df['parent_incident_type'] == 'Traffic'
traffic_df =  results_df[traffic_mask]
```
